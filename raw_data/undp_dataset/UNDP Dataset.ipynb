```python
json = r.json()
```

```python
json[2000]
```

```python
df = pd.DataFrame (json)
df.columns = ['indicator_id','indicator_name','country_code' , 'country', 'date', 'value']
df.head()
```

```python
# flipping the dataset
#df['indicator'] = df.indicator_id +" : "+ df.indicator_name
df['indicator'] = df.indicator_name
df.drop (columns=['indicator_id','indicator_name'], inplace = True)
undp = df.pivot_table(values='value', index=['country_code', 'country', 'date'], columns='indicator', aggfunc='first').reset_index()
undp.head()
```

```python
undp.to_csv ('undp_output.csv', index=False)
```
